# Plotting the results of test_training_set_size.py

```python
import os
import logging
import numpy as np
import h5py

from skqd_z2lgt.parameters import Parameters
from skqd_z2lgt.tasks.train_generator import train_for
from skqd_z2lgt.tasks.preprocess import load_reco

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger()
```

```python
sizes = [100_000, 200_000, 500_000, 800_000, 950_000]
```

```python
# Results from test_train_set_size.py
istep = 1
records = {}
for num_train in sizes:
    records[num_train] = {}
    with h5py.File(f'/data/iiyama/2dz2/train_test/ik{istep}_n{num_train}.h5', 'r', libver='latest') as source:
        for key, value in source['records'].items():
            records[num_train][key] = value[()]
```

```python
print(records[100000].keys())
```

```python
import matplotlib.pyplot as plt

for size in sizes:
    plt.plot(records[size]['test_success_100'], label=f'{size}')

plt.legend();
```
